```python
week_days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]

time_slots = (
            [{"day": day, "time": "08:30-09:50"} for day in week_days] + 
            [{"day": day, "time": "10:00-11:20"} for day in week_days] + 
            [{"day": day, "time": "11:30-12:50"} for day in week_days] + 
            [{"day": day, "time": "01:00-02:20"} for day in week_days] +
            [{"day": day, "time": "02:30-03:50"} for day in week_days] +
            [{"day": day, "time": "03:55-05:15"} for day in week_days] +
            [{"day": day, "time": "05:20-06:40"} for day in week_days] +
            [{"day": day, "time": "06:45-08:05"} for day in week_days])*2



```

```python
import random
import time

def time_overlaps(time1, time2):
    start1, end1 = time1.split('-')
    start2, end2 = time2.split('-')

    start1_minutes = int(start1.split(':')[0]) * 60 + int(start1.split(':')[1])
    end1_minutes = int(end1.split(':')[0]) * 60 + int(end1.split(':')[1])
    start2_minutes = int(start2.split(':')[0]) * 60 + int(start2.split(':')[1])
    end2_minutes = int(end2.split(':')[0]) * 60 + int(end2.split(':')[1])

    return max(start1_minutes, start2_minutes) < min(end1_minutes, end2_minutes)


def select_room(course, available_rooms):
    for room in available_rooms:
        if course['Type'] == 'Theory' and room['type_of_room'] == 'classroom' and room['capacity'] >= course['ClassStrength']:
            return room['name']
        elif course['Type'] == 'Lab' and room['type_of_room'] == 'large hall' and room['capacity'] >= course['ClassStrength']:
            return room['name']
    return None



def valid_time_slot(course, time):
    if course['Type'] == 'Theory' and time['time'] in ["08:30-09:50", "10:00-11:20", "11:30-12:50", "01:00-02:20", "02:30-03:50"]:
        return True
    elif course['Type'] == 'Lab' and time['time'] in ["03:55-05:15", "05:20-06:40", "06:45-08:05"]:
        return True
    return False

def prev_day(day):
    day_index = week_days.index(day)
    return week_days[day_index - 1] if day_index > 0 else None

def next_day(day):
    day_index = week_days.index(day)
    return week_days[day_index + 1] if day_index < len(week_days) - 1 else None


def valid_day_pair(assigned_days, new_day, allowed_day_pairs):
    if not assigned_days:  
        return any(new_day in pair for pair in allowed_day_pairs)
    elif len(assigned_days) == 1:  
        assigned_day = next(iter(assigned_days))
        return any({assigned_day, new_day} == pair for pair in allowed_day_pairs)
    return False 

def find_consecutive_time_slots(day, time):
        """Find consecutive time slots based on a starting time slot."""
        start_index = time_slots.index(time)
        # Ensure the next time slot is on the same day and is consecutive
        if start_index + 1 < len(time_slots) and time_slots[start_index + 1]['day'] == day:
            return time_slots[start_index + 1]
        return None

    
#                     if course['Section'] in section_classroom:
#                         assigned_room = section_classroom[course['Section']]
#                         if room['name'] != assigned_room:
#                             continue  # Skip if a different room is assigned for the section
#                     else:
#                         section_classroom[course['Section']] = room['name']  # Assign the room for the section

def create_individual(rooms, courses, time_slots):
    timetable = []
    room_availability = {}
    section_course_assigned_days = {}  
    course_day_usage = {} 
    professor_course_count = {}
    section_course_count = {} 
    section_time_slots = {} 

    for course in courses:
        assigned_days = section_course_assigned_days.get((course['Section'], course['Code']), set())

        if professor_course_count.get(course['Course Instructor'], 0) >= 3:
            continue 

        if section_course_count.get(course['Section'], 0) >= 5:
            continue  

        for time in time_slots:
            if valid_time_slot(course, time) and not course_day_usage.get((course['Code'], course['Section'], time['day'])):
                if section_time_slots.get((course['Section'], time['day'], time['time']), False):
                    continue  
                    
                room = select_room(course, [room for room in rooms if time['time'] not in room_availability.get(room['name'], [])])
                if room and not any(s['Instructor'] == course['Course Instructor'] and s['Day'] == time['day'] and s['Time'] == time['time'] for s in timetable):
                    if course['Type'] == 'Lab':
                        if len(assigned_days) >= 1:
                            continue

                        next_time = find_consecutive_time_slots(time['day'], time)
                        if next_time and not any(s['Instructor'] == course['Course Instructor'] and s['Day'] == time['day'] and s['Time'] in [time['time'], next_time['time']] for s in timetable):
                            timetable.extend([
                                {'Course Code': course['Code'], 'Section': course['Section'], 'Instructor': course['Course Instructor'], 'Room': room, 'Day': time['day'], 'Time': time['time'], 'Type': course['Type']},
                                {'Course Code': course['Code'], 'Section': course['Section'], 'Instructor': course['Course Instructor'], 'Room': room, 'Day': time['day'], 'Time': next_time['time'], 'Type': course['Type']}
                            ])
                            room_availability.setdefault(room['name'], []).extend([time['time'], next_time['time']])
                            assigned_days.add(time['day'])
                            section_time_slots[(course['Section'], time['day'], time['time'])] = True
                            section_time_slots[(course['Section'], time['day'], next_time['time'])] = True
                            continue

                    if course['Type'] == 'Theory':
                        if len(assigned_days) >= 2:
                            continue
                            
                        allowed_day_pairs = [
                            {'Monday', 'Wednesday'}, 
                            {'Monday', 'Thursday'}, 
                            {'Monday', 'Friday'},
                            {'Tuesday', 'Thursday'},
                            {'Tuesday', 'Friday'},
                            {'Wednesday', 'Friday'}
                        ]

                        if time['day'] in assigned_days or any(abs(week_days.index(day) - week_days.index(time['day'])) == 1 for day in assigned_days):
                            continue

                        if not valid_day_pair(assigned_days, time['day'], allowed_day_pairs):
                            continue

                    timetable.append({
                        'Course Code': course['Code'],
                        'Section': course['Section'],
                        'Instructor': course['Course Instructor'],
                        'Room': room,
                        'Day': time['day'],
                        'Time': time['time'],
                        'Type': course['Type']
                    })
                    professor_course_count[course['Course Instructor']] = professor_course_count.get(course['Course Instructor'], 0) + 1
                    section_course_count[course['Section']] = section_course_count.get(course['Section'], 0) + 1
                    room_availability.setdefault(room, []).append(time['time'])
                    assigned_days.add(time['day'])
                    section_course_assigned_days[(course['Section'], course['Code'])] = assigned_days
                    course_day_usage[(course['Code'], course['Section'], time['day'])] = True
                    section_time_slots[(course['Section'], time['day'], time['time'])] = True

    return timetable


def calculate_fitness(individual):
    fitness = 50  
    instructor_times = {}
    room_times = {}
    section_course_times = {}

    for session in individual:
        instructor_key = (session['Instructor'], session['Day'], session['Time'])
        room_key = (session['Room'], session['Day'], session['Time'])
        section_course_key = (session['Section'], session['Course Code'], session['Type'])

        if instructor_key in instructor_times:
            fitness -= 10
        else:
            instructor_times[instructor_key] = True

        if room_key in room_times:
            fitness -= 10
        else:
            room_times[room_key] = True

        if section_course_key in section_course_times:
            section_course_times[section_course_key].append(session['Day'])
        else:
            section_course_times[section_course_key] = [session['Day']]

    for key, days in section_course_times.items():
        if 'Lab' in key[2] and len(set(days)) != 1:
            fitness -= 5
        if 'Theory' in key[2] and len(set(days)) != 2:
            fitness -= 5

    return fitness

def mutate(individual, time_slots, mutation_rate=0.05):
    for session in individual:
        if random.random() < mutation_rate:
            valid_times = [time for time in time_slots if valid_time_slot(session, time)]
            if valid_times:
                new_time = random.choice(valid_times)
                if session['Type'] == 'Theory' and new_time['time'] not in ["08:30-09:50", "10:00-11:20", "11:30-12:50"]:
                    # Ensure theory sessions are in the morning
                    new_time = random.choice([time for time in valid_times if time['time'] in ["08:30-09:50", "10:00-11:20", "11:30-12:50"]])
                elif session['Type'] == 'Lab' and new_time['time'] not in ["03:55-05:15", "05:20-06:40", "06:45-08:05"]:
                    # Ensure lab sessions are in the afternoon
                    new_time = random.choice([time for time in valid_times if time['time'] in ["03:55-05:15", "05:20-06:40", "06:45-08:05"]])
                session['Day'] = new_time['day']
                session['Time'] = new_time['time']
    return individual

def selection(population, tournament_size=5):
    selected = []
    for _ in range(len(population)):
        tournament = random.sample(population, tournament_size)
        best_individual = max(tournament, key=calculate_fitness)
        selected.append(best_individual)
    return selected

def crossover(parent1, parent2):
    crossover_point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def genetic_algorithm(rooms, courses, time_slots, population_size=100, generations=200, mutation_rate=0.05):
    population = [create_individual(rooms, courses, time_slots) for _ in range(population_size)]
    for generation in range(generations):
        start_time = time.time()
        population = sorted(population, key=lambda x: calculate_fitness(x), reverse=True)
        new_population = selection(population)
        next_generation = []
        while len(next_generation) < population_size:
            parent1, parent2 = random.sample(new_population, 2)
            child1, child2 = crossover(parent1, parent2)
            next_generation.append(mutate(child1, time_slots, mutation_rate))
            if len(next_generation) < population_size:
                next_generation.append(mutate(child2, time_slots, mutation_rate))
        population = next_generation
        end_time = time.time()
        
        # print(f"Generation {generation + 1} completed in {end_time - start_time} seconds.")

    #population = lab_time_checking(population)
    return max(population, key=lambda x: calculate_fitness(x))
```

```python
def print_timetable(timetable):
    # Sort the timetable by day and then by time for readability
    sorted_timetable = sorted(timetable, key=lambda x: (week_days.index(x['Day']), x['Time']))
    
    # Print each session in the specified format
    for session in sorted_timetable:
        print(f"{session['Day']} - {session['Room']} - {session['Course Code']} - {session['Section']} - {session['Type']} - {session['Instructor']} - {session['Time']}")

# Execute the genetic algorithm
best_timetable = genetic_algorithm(rooms, courses, time_slots)
print_timetable(best_timetable)
```
